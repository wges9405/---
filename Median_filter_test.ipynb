```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from scipy import signal
```

```python
body_acc_x = pd.read_csv('./body_acc_x_test.txt', delim_whitespace=True, header=None)[0].to_numpy()
```

```python
def showplot(data):
    plt.plot(data)
    plt.show()
    plt.clf()
def filter(data):
    return signal.medfilt(data)
```
