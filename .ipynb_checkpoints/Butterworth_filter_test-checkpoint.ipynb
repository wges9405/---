```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import signal
```

```python
body_acc_x = pd.read_csv('./body_acc_x_test.txt', delim_whitespace=True, header=None)[0].to_numpy()
body_acc_y = pd.read_csv('./body_acc_y_test.txt', delim_whitespace=True, header=None)[0].to_numpy()
body_acc_z = pd.read_csv('./body_acc_z_test.txt', delim_whitespace=True, header=None)[0].to_numpy()
```

```python
def showplot(data):
    plt.plot(data)
    plt.show()
    plt.clf()
def third_order_low_butterworth_filter_D(cutoff_freq, sensor_freq, data):
    b, a = signal.butter(3, 2*cutoff_freq/sensor_freq, btype='lowpass', analog=False, output='ba')
    filtedData = signal.filtfilt(b, a, data)
    return filtedData
```

```python
def third_order_low_butterworth_filter_A(cutoff_freq, sensor_freq, data):
    b, a = signal.butter(3, 2*cutoff_freq/sensor_freq, btype='lowpass', analog=False, output='ba')
    filtedData = signal.filtfilt(b, a, data)
    return filtedData
```

```python
sensor_freq = 50 #Hz
cutoff_freq = 0.3 #Hz
```

```python
filted_body_acc_x_D = third_order_low_butterworth_filter_D(0.3, 50, body_acc_x)
filted_body_acc_x_A = third_order_low_butterworth_filter_A(20, 50, body_acc_x)
filted_body_acc_y_D = third_order_low_butterworth_filter_D(0.3, 50, body_acc_y)
filted_body_acc_y_A = third_order_low_butterworth_filter_A(20, 50, body_acc_y)
filted_body_acc_z_D = third_order_low_butterworth_filter_D(0.3, 50, body_acc_z)
filted_body_acc_z_A = third_order_low_butterworth_filter_A(20, 50, body_acc_z)
```

```python
showplot(body_acc_x)
showplot(filted_body_acc_x_D)
showplot(filted_body_acc_x_A)
```

```python
showplot(body_acc_x)
showplot(filted_body_acc_y_D)
showplot(filted_body_acc_y_A)
```

```python
showplot(body_acc_x)
showplot(filted_body_acc_z_D)
showplot(filted_body_acc_z_A)
```
