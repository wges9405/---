```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from scipy import signal
```

```python
body_acc_x = pd.read_csv('./body_acc_x_test.txt', delim_whitespace=True, header=None)[0].to_numpy()
```

```python
def showplot(data):
    plt.plot(data)
    plt.show()
    plt.clf()
def filter(cutoff_freq, sensor_freq, analog, data):
    b, a = signal.butter(3, 2*cutoff_freq/sensor_freq, btype='lowpass', analog=analog, output='ba')
    filtedData = signal.filtfilt(b, a, data)
    return filtedData
```

```python
showplot(filter(0.1, 50, False, body_acc_x))
showplot(filter(0.1, 50, True, body_acc_x))
```

```python
showplot(filter(0.3, 50, False, body_acc_x))
showplot(filter(0.3, 50, True, body_acc_x))
```
