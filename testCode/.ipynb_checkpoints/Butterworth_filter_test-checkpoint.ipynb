```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from scipy import signal
```

```python
body_acc_x = pd.read_csv('../testData/body_acc_x_test.txt', delim_whitespace=True, header=None).to_numpy()
y = pd.read_csv('../testData/y_test.txt', delim_whitespace=True, header=None).to_numpy()

ACTIVITIES = {
    0: 'WALKING',
    1: 'WALKING_UPSTAIRS',
    2: 'WALKING_DOWNSTAIRS',
    3: 'SITTING',
    4: 'STANDING',
    5: 'LAYING',
}
```

```python
def showplot(data, after):
    plt.plot(data)
    plt.plot(after)
    plt.show()
def filter(cutoff_freq, sensor_freq, analog, data):
    b, a = signal.butter(3, 2*cutoff_freq/sensor_freq, btype='lowpass', analog=analog, output='ba')
    filtedData = signal.filtfilt(b, a, data)
    return filtedData
```

```python
index = random.randint(0,2947)
data = body_acc_x[index]
print('index: ',index, '\ty: ', y[index][0], '\tActivity: ', ACTIVITIES[y[index][0]-1], '\n')

for int in np.linspace(0.1,1,10):
    print('cutoff_freq: ',int)
    after = filter(int, 50, False, body_acc_x)[index]
    showplot(data, after)
#     print('diff: \n', after-data)
    print('diff. mean: ', np.mean(after-data))
    print('diff. std: ', np.std(after-data), '\n')
for int in range(1, 25, 1):
    print('cutoff_freq: ',int)
    after = filter(int, 50, False, body_acc_x)[index]
    showplot(data, after)
#     print('diff: \n', after-data)
    print('diff. mean: ', np.mean(after-data))
    print('diff. std: ', np.std(after-data), '\n')
```
