```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from scipy import signal
```

```python
body_acc_x = pd.read_csv('../testData/body_acc_x_test.txt', delim_whitespace=True, header=None).to_numpy()
y = pd.read_csv('../testData/y_test.txt', delim_whitespace=True, header=None).to_numpy()
```

```python
def showplot(data, after):
    plt.plot(data)
    plt.plot(after, 'r')
    plt.show()
def filter(data, sliding_window):
    return signal.medfilt(data, sliding_window)
```

```python
index = random.randint(0,2947)
data = body_acc_x[index]
sliding_window = [3]

print('index: ',index, '\ty: ', y[index][0], '\n')

for int in range(1, 16, 2):
    print('sliding window size: ',int)
    after = filter(body_acc_x, [1,int])[index]
    showplot(data, after)
    #     print('diff: \n', after-data)
    print('diff. mean: ', np.mean(after-data))
    print('diff. std: ', np.std(after-data), '\n')
```
